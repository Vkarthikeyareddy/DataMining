```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
import os

os.chdir('C:\\Users\\manis\\OneDrive\\Desktop\\545 Data Mining Bus Intell\\Lab 20')
```

```python
market_basket_data = pd.read_csv('marketBasketAnalysis.csv')
```

```python
# Selecting relevant columns for market basket analysis
mba_data = market_basket_data[['Order ID', 'SKU']]

# Grouping the data by 'Order ID' and aggregating the 'SKU' values into lists
transactions = mba_data.groupby('Order ID')['SKU'].apply(list)
```

```python
# display the first 5 rows of mba_data
print(mba_data.head())
```

```python
# display the first 5 rows of the transactions dataset
transactions.head()
```

```python
# Transforming the data into a one-hot encoded boolean array

te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)

```

```python
# display the first 5 rows of te_ary

print(te_ary[:5])
```

```python
# how many columns are there in te_ary?
print(len(te_ary[0]))
```

```python
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
frequent_itemsets = apriori(df, min_support=0.30, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.2)
print(rules.head())
```

```python
# print the number of rules generated
print(len(rules))
```

```python
# print the first 20 rules
rules.head(20)

```

```python
# arrange the rules in descending order of confidence
rules.sort_values('confidence', ascending = False, inplace = True)
rules.head(20)
```
